```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colorbar as colorbar
from cartopy import crs
#from mpl_toolkits.Basemap import Basemap
import matplotlib.colors as mcolors
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeat
from matplotlib import gridspec
from scipy.io import loadmat

```

```python
#Chargement des donnees a partir de notre de reportoire et leur stockage dans un dictionnaire python
comp_U=loadmat('C:/Users/USER-PC/Documents/Python/matlab/J1/data1/M2_comp_U.mat')
lat=loadmat('C:/Users/USER-PC/Documents/Python/matlab/J1/data1/M2_lat.mat')
lon=loadmat('C:/Users/USER-PC/Documents/Python/matlab/J1/data1/M2_lon.mat')
```

```python
# Extraction des données presentes dans le dicionnaire python
U = comp_U['comp_U']  
la = lat['lat']            
lo = lon['lon']            
```

```python
# Affichage des figures
fig=plt.figure()
ax1 =fig.add_subplot(111, projection=ccrs.PlateCarree())
#ax1.coastlines()
ax1.add_feature(cfeat.NaturalEarthFeature('physical', 'land', '10m', edgecolor='black', facecolor='gray', linewidth=2))
ax1.add_feature(cfeat.BORDERS, zorder=1e4,linewidth=0.5,edgecolor='darkgray')

X, Y = np.meshgrid(lo, la)
plot_U= ax1.pcolor(X, Y, U, cmap='RdBu_r')

cbar = fig.colorbar(plot_U, ax=ax1, orientation='vertical', label='courant (cm/s)', shrink=0.8)
cbar.set_ticks([-160, -130, -90, -60, -30, 0, 30, 60, 90, 120, 160])
# Définir les axes
ax1.set_xlabel('Longitude')
ax1.set_ylabel('Latitude')
ax1.set_title('Courant zonal U')
#plt.grid()
plt.tight_layout()
plt.show()
```
